```python
import numpy as np

matrixC = np.random.randint(0,10,size=(4,3))


print("Random Matrix C : \n", matrixC)


matrixCT = matrixC.transpose()

print("Transpose of Matrix C : \n",matrixCT)

matrixA1 = np.dot(matrixCT,matrixC)
print("Matrix A : \n",matrixA1)

eigenvalues, eigenvectors = np.linalg.eig(matrixA1)

print("Eigen values : ",np.round(eigenvalues,5))

print("Eigen Vector : \n",np.round(eigenvectors,5))

lambdaValues = np.poly(matrixA1)
print("lambda values : ",lambdaValues)

charEq = np.poly1d(lambdaValues)

print("characteristic equation : \n",charEq)
```

```python
import numpy as np

def power_method(matrix, max_iterations=10):
    n = matrix.shape[0]
    
    # Initialize a random vector as the initial guess for the eigenvector
    # x = np.random.rand(n)
    x = np.ones(n,dtype=int)
    
    for i in range(max_iterations):
        # Power iteration step
        y = np.dot(matrix, x)
        # print(f"x : { x }  &  y(matrix and x) dot product = ",y)

        # Extract the largest magnitude component of y
        eigenvalue = np.max(y)
        
        # Normalize the eigenvector
        x = y / eigenvalue
        
        # Print the current iterate and eigenvalue
        print(f"Iteration {i + 1}: Eigenvalue = {eigenvalue:.5f} X = {x}")
    
    # Return the largest eigenvalue and the corresponding eigenvector
    return eigenvalue, x

# Using the previously generated matrix A1
# A1 = np.array([[30, 15, 26], [15, 19, 18], [26, 18, 29]])
# print("A1 = ",matrixA1)

np.set_printoptions(precision=5, suppress=True)

# Run the power method
lambda1, x1 = power_method(matrixA1)


# Display the first 10 iterates of the eigenvalue
# print("\nFirst 10 iterates of eigenvalue:")
# print(lambda1)

# Normalize the corresponding eigenvector outside the function
x1_hat = x1 / np.linalg.norm(x1, 2)

# Display the final results
print("\nFinal Results:")
print(f"Largest Eigenvalue (λ1): {lambda1:.5f}")
print(f"Corresponding Eigenvector (x1): {x1}")
print(f"Normalized eigenvector (ˆx1): {x1_hat}")


```

```python

# Construct matrix A2 = (A1 - ˆx1ˆxT1A1)
A2 = matrixA1 - np.outer(x1_hat, x1_hat) @ matrixA1

np.set_printoptions(precision=5, suppress=True)

# print("A2 : ",A2)
# Run the power method on A2
lambda_2, x2 = power_method(A2)

x2_hat = x2 / np.linalg.norm(x2, 2)
# Display the results for A2
print("\nFinal Results:")
print(f"Largest Eigenvalue (λ1): {lambda_2:.5f}")
print(f"Corresponding Eigenvector (x2): {x2}")
print(f"Normalized eigenvector (ˆx2): {x2_hat}")
```

```python


# Construct matrix A3 = (A1 - ˆx1ˆxT1A1 - ˆx2ˆxT2A1)
A3 = matrixA1 - np.outer(x1_hat, x1_hat) @ matrixA1 - np.outer(x2_hat, x2_hat) @ matrixA1

# print("A3 : ",A3)

# Run the power method on A3
lambda_3, x3 = power_method(A3)

x3_hat = x3 / np.linalg.norm(x3, 2)

# Display the results for A3
print("\nFinal Results:")
print(f"Largest Eigenvalue (λ3): {lambda_3:.5f}")
print(f"Corresponding Eigenvector (x3): {x3}")
print(f"Normalized eigenvector (ˆx3): {x3_hat}")
```
